```python
!pip install -q timm
!pip install -q torchinfo
!pip install -q seaborn

import os, random, math, time, shutil
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import transforms
from torchvision.datasets import FashionMNIST
import timm
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from torchinfo import summary

def set_seed(seed=0):
    random.seed(seed); np.random.seed(seed)
    torch.manual_seed(seed); torch.cuda.manual_seed_all(seed)
set_seed(0)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Device:", device)

# Cell 2: Hyperparams & paths
IMG_SIZE = 128
BATCH_SIZE = 192
NUM_WORKERS = 2
EPOCHS = 6   # requested
ROOT = "/content"
ARTIFACT_DIR = os.path.join(ROOT, "ectf_vae_artifacts")
os.makedirs(ARTIFACT_DIR, exist_ok=True)

ENSEMBLE_SIZE = 2
TOKEN_DIM = 128
NUM_HEADS = 4
NUM_LAYERS = 2
VAE_KL_BETA = 1e-3
VAE_RECON_LAMBDA = 1.0
LABEL_SMOOTHING = 0.1
LEARNING_RATE = 3e-4
WEIGHT_DECAY = 1e-4

def denormalize_img(img_tensor, mean=0.5, std=0.5):
    # img_tensor: torch tensor (3,H,W) in normalized [-1,1] range from Normalize(0.5,0.5)
    img = img_tensor.clone().cpu()
    img = img * std + mean   # reverse Normalize
    img = img.permute(1,2,0).numpy()
    img = np.clip(img, 0.0, 1.0)
    return img

train_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Lambda(lambda x: x.repeat(3,1,1)),
    transforms.Normalize([0.5]*3, [0.5]*3),
])

test_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Lambda(lambda x: x.repeat(3,1,1)),
    transforms.Normalize([0.5]*3, [0.5]*3),
])

train_ds = FashionMNIST(root='data', train=True, download=True, transform=train_transform)
test_ds  = FashionMNIST(root='data', train=False, download=True, transform=test_transform)
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
test_loader  = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

print("Train size:", len(train_ds), "Test size:", len(test_ds))

# quick sample display (uses denormalize)
def show_samples(dataset, n=6):
    fig, axs = plt.subplots(1,n,figsize=(12,3))
    for i in range(n):
        x,y = dataset[i]
        axs[i].imshow(denormalize_img(x))
        axs[i].set_title(str(y)); axs[i].axis('off')
    plt.show()

show_samples(train_ds, n=6)

class PositionalEncoding1D(nn.Module):
    def __init__(self, d_model, max_len=2000):
        super().__init__()
        pe = torch.zeros(max_len, d_model)
        pos = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(pos * div_term)
        pe[:, 1::2] = torch.cos(pos * div_term)
        self.register_buffer('pe', pe)
    def forward(self, x):
        T = x.size(1)
        return x + self.pe[:T].unsqueeze(0)

class TokenVAE(nn.Module):
    def __init__(self, token_dim=128, latent_dim=64, hidden_dim=128, H=8, W=8):
        super().__init__()
        self.token_dim = token_dim
        self.latent_dim = latent_dim
        self.encoder = nn.Sequential(
            nn.Linear(token_dim, hidden_dim),
            nn.GELU(),
            nn.Linear(hidden_dim, latent_dim*2)
        )
        self.decoder_fc = nn.Sequential(
            nn.Linear(latent_dim, hidden_dim),
            nn.GELU(),
            nn.Linear(hidden_dim, token_dim*H*W)
        )
        self.H, self.W = H, W
    def forward(self, tokens):
        pooled = tokens.mean(dim=1)  # (B,D)
        stats = self.encoder(pooled)
        mu, logvar = stats.chunk(2, dim=1)
        std = (0.5*logvar).exp()
        eps = torch.randn_like(std)
        z = mu + eps * std
        dec = self.decoder_fc(z)
        dec = dec.view(-1, self.token_dim, self.H, self.W)
        kld = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp(), dim=1).mean()
        return dec, kld

class ECTF_VAE(nn.Module):
    def __init__(self, backbone_name='efficientnet_b0', pretrained=True, num_classes=10, token_dim=128, num_heads=4, num_layers=2, vae_latent=64, dropout=0.15):
        super().__init__()
        self.backbone = timm.create_model(backbone_name, pretrained=pretrained, features_only=True, out_indices=[4])
        # sample to get channels and spatial dims
        with torch.no_grad():
            sample = self.backbone(torch.randn(1,3,IMG_SIZE,IMG_SIZE))
            f = sample[-1] if isinstance(sample,(list,tuple)) else sample
            channels = f.shape[1]; _,_,H,W = f.shape
        self.token_proj = nn.Conv2d(channels, token_dim, kernel_size=1)
        self.H, self.W = H, W
        encoder_layer = nn.TransformerEncoderLayer(d_model=token_dim, nhead=num_heads, dim_feedforward=token_dim*2, dropout=dropout, activation='gelu', batch_first=True)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.pos_enc = PositionalEncoding1D(token_dim, max_len=H*W+10)
        self.token_vae = TokenVAE(token_dim, latent_dim=vae_latent, hidden_dim=vae_latent*2, H=H, W=W)
        self.global_pool = nn.AdaptiveAvgPool2d(1)
        fusion_dim = token_dim + channels
        self.classifier = nn.Sequential(
            nn.Linear(fusion_dim, fusion_dim//2),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(fusion_dim//2, num_classes)
        )
    def forward(self, x, return_tok_recon=False):
        feats = self.backbone(x)
        f = feats[-1] if isinstance(feats,(list,tuple)) else feats
        pool = self.global_pool(f).view(f.size(0), -1)
        t = self.token_proj(f)
        B,D,H,W = t.shape
        tokens = t.view(B, D, H*W).permute(0,2,1)
        tokens = self.pos_enc(tokens)
        t_out = self.transformer(tokens)
        tpool = t_out.mean(dim=1)
        recon_map, kld = self.token_vae(t_out)
        fused = torch.cat([tpool, pool], dim=1)
        logits = self.classifier(fused)
        if return_tok_recon:
            return logits, recon_map, kld
        return logits, recon_map, kld

# Print model summary for quick check (optional)
# model_check = ECTF_VAE(pretrained=True, token_dim=TOKEN_DIM, num_heads=NUM_HEADS, num_layers=NUM_LAYERS).to(device)
# summary(model_check, input_size=(1,3,IMG_SIZE,IMG_SIZE))

def classification_loss_fn(logits, labels, smoothing=0.0):
    if smoothing > 0.0:
        n_classes = logits.size(1)
        with torch.no_grad():
            true_dist = torch.zeros_like(logits)
            true_dist.fill_(smoothing / (n_classes - 1))
            true_dist.scatter_(1, labels.data.unsqueeze(1), 1.0 - smoothing)
        loss = torch.mean(torch.sum(-true_dist * F.log_softmax(logits, dim=1), dim=1))
    else:
        loss = F.cross_entropy(logits, labels)
    return loss

def train_one_epoch(model, loader, optimizer, device, smoothing=0.0, vae_recon_lambda=1.0, vae_kl_beta=1e-3):
    model.train()
    running_loss = running_ce = running_recon = running_kld = 0.0
    total = 0
    preds = []; trues = []
    for xb, yb in tqdm(loader, desc="Train", leave=False):
        xb, yb = xb.to(device), yb.to(device)
        logits, recon_map, kld = model(xb, return_tok_recon=True)
        ce = classification_loss_fn(logits, yb, smoothing)
        with torch.no_grad():
            feats = model.backbone(xb); f = feats[-1] if isinstance(feats,(list,tuple)) else feats
            target = model.token_proj(f)
        recon_loss = F.mse_loss(recon_map, target)
        loss = ce + vae_recon_lambda * recon_loss + vae_kl_beta * kld
        optimizer.zero_grad(); loss.backward(); optimizer.step()
        n = xb.size(0)
        running_loss += loss.item() * n
        running_ce += ce.item() * n
        running_recon += recon_loss.item() * n
        running_kld += kld.item() * n
        total += n
        preds.append(logits.argmax(dim=1).detach().cpu().numpy())
        trues.append(yb.detach().cpu().numpy())
    preds = np.concatenate(preds); trues = np.concatenate(trues)
    acc = (preds == trues).mean()
    return running_loss/total, running_ce/total, running_recon/total, running_kld/total, acc

def evaluate(model, loader, device):
    model.eval()
    preds = []; trues = []
    with torch.no_grad():
        for xb, yb in loader:
            xb = xb.to(device)
            logits, _, _ = model(xb, return_tok_recon=True)
            preds.append(logits.argmax(dim=1).cpu().numpy())
            trues.append(yb.numpy())
    preds = np.concatenate(preds); trues = np.concatenate(trues)
    acc = (preds==trues).mean()
    f1 = f1_score(trues, preds, average='macro')
    return acc, f1, preds, trues

def ensemble_predict(models, loader, device):
    for m in models: m.eval()
    preds = []; trues = []
    with torch.no_grad():
        for xb, yb in loader:
            xb = xb.to(device)
            logits_sum = None
            for m in models:
                logits, _, _ = m(xb, return_tok_recon=False)
                if logits_sum is None: logits_sum = logits
                else: logits_sum = logits_sum + logits
            avg_logits = logits_sum / len(models)
            preds.append(avg_logits.argmax(dim=1).cpu().numpy())
            trues.append(yb.numpy())
    preds = np.concatenate(preds); trues = np.concatenate(trues)
    acc = (preds==trues).mean()
    f1 = f1_score(trues, preds, average='macro')
    return acc, f1, preds, trues

# Safe load helper: handles older full checkpoints by allowlisting numpy scalar then resaving state_dict
def safe_load_checkpoint(path, map_location='cpu'):
    try:
        chk = torch.load(path, map_location=map_location)
        if isinstance(chk, dict):
            # if it's a full checkpoint with 'model_state' key, return it
            if 'model_state' in chk:
                return chk
            # if it's a state_dict directly, wrap it
            return {'model_state': chk}
        else:
            # unexpected type
            return {'model_state': chk}
    except Exception as e:
        # try allowlist numpy scalar (trusted source)
        import numpy as np
        try:
            torch.serialization.add_safe_globals([np._core.multiarray.scalar])
            chk = torch.load(path, map_location=map_location)
            if isinstance(chk, dict) and 'model_state' in chk:
                return chk
            return {'model_state': chk}
        except Exception as e2:
            raise RuntimeError(f"Failed to load checkpoint {path}: {e2}")

def train_and_save_model(model_idx, seed=0, epochs=EPOCHS, save_dir=ARTIFACT_DIR):
    set_seed(seed)
    print(f"\n--> Training model {model_idx} with seed {seed}")
    model = ECTF_VAE(pretrained=True, token_dim=TOKEN_DIM, num_heads=NUM_HEADS, num_layers=NUM_LAYERS, vae_latent=64).to(device)
    optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=3, gamma=0.5)
    best_val = 0.0
    model_path = os.path.join(save_dir, f"ectf_vae_model_{model_idx}.pth")
    state_path = os.path.join(save_dir, f"ectf_vae_model_{model_idx}_state_dict.pth")
    for epoch in range(1, epochs+1):
        t0 = time.time()
        train_loss, train_ce, train_recon, train_kld, train_acc = train_one_epoch(model, train_loader, optimizer, device, smoothing=LABEL_SMOOTHING, vae_recon_lambda=VAE_RECON_LAMBDA, vae_kl_beta=VAE_KL_BETA)
        val_acc, val_f1, _, _ = evaluate(model, test_loader, device)
        scheduler.step()
        print(f"Model {model_idx} Epoch {epoch}/{epochs} | Train loss {train_loss:.4f} CE {train_ce:.4f} Recon {train_recon:.4f} KLD {train_kld:.6f} TrainAcc {train_acc:.4f} || ValAcc {val_acc:.4f} ValF1 {val_f1:.4f} | time {time.time()-t0:.1f}s")
        if val_acc > best_val:
            best_val = val_acc
            # save full checkpoint (for reproducibility) and also a safe state_dict copy
            torch.save({'model_state': model.state_dict(), 'epoch': epoch, 'val_acc': val_acc}, model_path)
            torch.save(model.state_dict(), state_path)
            print("Saved best full checkpoint ->", model_path)
            print("Saved safe state_dict ->", state_path)
    return model_path, state_path

# Train ensemble members
saved_full = []
saved_state = []
seeds = [42, 999]  # different seeds
for i in range(ENSEMBLE_SIZE):
    p_full, p_state = train_and_save_model(i, seed=seeds[i], epochs=EPOCHS, save_dir=ARTIFACT_DIR)
    saved_full.append(p_full); saved_state.append(p_state)

print("\nSaved checkpoints:")
for a,b in zip(saved_full, saved_state):
    print(a, " | ", b)

models = []
for state_path in saved_state:
    state = torch.load(state_path, map_location=device)
    m = ECTF_VAE(pretrained=True, token_dim=TOKEN_DIM, num_heads=NUM_HEADS, num_layers=NUM_LAYERS, vae_latent=64).to(device)
    m.load_state_dict(state)
    models.append(m)
    print("Loaded state_dict:", state_path)

ens_acc, ens_f1, ens_preds, ens_trues = ensemble_predict(models, test_loader, device)
print(f"\nEnsemble result: Acc {ens_acc:.4f} | Macro F1 {ens_f1:.4f}")
print(classification_report(ens_trues, ens_preds, digits=4))

# Confusion matrix plot
cm = confusion_matrix(ens_trues, ens_preds)
plt.figure(figsize=(8,6)); sns.heatmap(cm, annot=True, fmt='d', cmap='Blues'); plt.title("Ensemble Confusion Matrix"); plt.show()

summary_txt = f"""ECTF + Token-VAE ensemble artifacts
Ensemble size: {ENSEMBLE_SIZE}
Epochs per member: {EPOCHS}
Image size: {IMG_SIZE}
Token dim: {TOKEN_DIM}
Saved artifacts directory: {ARTIFACT_DIR}
"""
```
